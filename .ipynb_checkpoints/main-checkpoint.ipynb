```python
from string import Template

EMPLOYEE_CONTRACT_TYPE = "dataset/courier_contract_employee.txt"
FREELANCE_CONTRACT_TYPE = "dataset/courier_contract_freelance.txt"

def get_contract(contract_type, contract_date, courier_name, courier_address):
    with open(contract_type, 'r') as f_out: 
        template = f_out.read()

        template_vars = {
            "CONTRACT_DATE": contract_date,
            "COURIER_NAME": courier_name,
            "COURIER_ADDRESS": courier_address
        }
        
        return Template(template).safe_substitute(template_vars)
       

print(get_employee_contract("30.02.2025", "John Doe", "22 Maril street"))
```
